```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'NanumGothic'
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# 데이터 불러오기
network_log = pd.read_csv('network_log.csv')
network_log
```

```python
# 데이터 한글화 진행
network_log.rename(columns={
    'IP_INFO':'아이피주소',
    'STATUS_CONF':'정상/비정상',
    'COUNTRY':'접속국가',
    'PL_NAME':'접속자',
    'Access Time':'접속빈도'
}, inplace=True)
network_log = network_log[['접속자', '접속국가', '아이피주소', '접속빈도', '정상/비정상']]
network_log
```

```python
network_log['정상/비정상'].unique()
```

#### 타깃 데이터 분류!
정상적인 데이터의 분류는 validated와 benign으로 두었습니다. 
먼저 validated는 검증된 데이터를 의미하고 benign은 시스템이나 데이터에 해를 끼치지 않는 소프트웨어 활동을 의미하기 때문입니다.
비정상적인 데이터의 분류로는 Not Enough Data와 Pending으로 설정했습니다.
Not Enough Data의 경우 결론을 내리기에 데이터가 부족하다는 의미인데, 우회 기법을 통해서 자신을 은닉한 채로 패킷을 전달했을 가능성을 두었고 Pending의 경우 여러 요청이 들어올 때 잠시 멈춰있는 상태인데, DDOS 공격을 통한 접근일 가능성을 고려하여 이 두개의 값을 비정상적인 데이터로 설정했습니다!

```python
# 데이터의 라벨화
network_log['정상/비정상'] = network_log['정상/비정상'].map({'validated':0, 'benign':0,
                                                      'not_enough_data':1, 'pending':1})
network_log
```

단계 1. 데이터 분석하기

```python
# 결측치 확인
network_log.isnull().sum()
# 결측치가 접속자에 하나 있으므로 해당 행 drop
network_log.dropna(inplace=True)
# 다시 결측치 확인
network_log.isnull().sum()
```

```python
# 가장 먼저 어떤 접속자가 많은 접속을 시도하는지를 살펴보겠습니다.
self_best_access = network_log.groupby(['접속자', '정상/비정상'], as_index=False)['아이피주소'].count().sort_values(by='아이피주소', ascending=False)
self_best_access.rename(columns={'아이피주소':'접속횟수'}, inplace=True)
self_best_access
# 각 접속 횟수를 바탕으로 정상/비정상일때의 합성 그래프 도식화
fig, ax = plt.subplots(figsize=(15, 8))
ax.barh(self_best_access[self_best_access['정상/비정상']==1]['접속자'].head(10), 
         self_best_access[self_best_access['정상/비정상']==1]['접속횟수'].head(10), label='비정상')
ax.barh(self_best_access[self_best_access['정상/비정상']==0]['접속자'].head(10), 
         self_best_access[self_best_access['정상/비정상']==0]['접속횟수'].head(10), label='정상')
plt.legend()
plt.tight_layout()
plt.show()
```

위의 자료를 바탕으로 특정 접속자의 경우 높은 확률로 비정상 데이터를 전송하고, Alpha Strike Laabs GmbH의 경우 거의 100% 확률로 정상적인 데이터를 전송한다는 것을 알 수 있습니다.

```python
# 다음으로 어떤 국가에서 많은 접속을 시도하는지를 살펴보겠습니다.
self_best_access = network_log.groupby(['접속국가', '정상/비정상'], as_index=False)['아이피주소'].count().sort_values(by='아이피주소', ascending=False)
self_best_access.rename(columns={'아이피주소':'접속횟수'}, inplace=True)
self_best_access
# 각 접속 횟수를 바탕으로 정상/비정상일때의 합성 그래프 도식화
fig, ax = plt.subplots(figsize=(15, 8))
ax.barh(self_best_access[self_best_access['정상/비정상']==1]['접속국가'].head(10), 
         self_best_access[self_best_access['정상/비정상']==1]['접속횟수'].head(10), label='비정상')
ax.barh(self_best_access[self_best_access['정상/비정상']==0]['접속국가'].head(10), 
         self_best_access[self_best_access['정상/비정상']==0]['접속횟수'].head(10), label='정상')
plt.legend()
plt.tight_layout()
plt.show()
```

위의 자료를 바탕으로 미국의 경우 높은 확률로 비정상 데이터를 전송하고, DE의 경우 거의 100% 확률로 정상적인 데이터를 전송한다는 것을 알 수 있습니다.

```python
# 문자열과 카테고리는 상관계수를 구할 수 없으니 카테고리형은 OneHotEncoder로, 수치형은 StandardScaler을 통해 표준화를 시키겠습니다.
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
trans = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['접속빈도']),
        ('cat', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), ['접속자', '접속국가', '아이피주소'])
    ], remainder='drop'
)
```

```python
# 그리고 원본 데이터에서 훈련 데이터와 테스트 데이터를 구분하겠습니다.
data, target = network_log.drop(columns='정상/비정상'), network_log['정상/비정상']
```

```python
# 이제 정규화시킨 데이터를 한번 볼까요?
transform_data = trans.fit_transform(data)
scaled_data = pd.DataFrame(transform_data,
             columns=['접속자', '접속국가', '아이피주소', '접속빈도'])
scaled_data
```

```python
# 특성들도 표준화 시켰으니 이번에는 각 특성에 따른 정상/비정상 데이터의 상관관계를 살펴볼까요?
corr = scaled_data['아이피주소'].corr(network_log['정상/비정상'])
print('아이피주소와 정상/비정상 열의 상관관계 :', corr)
corr = scaled_data['접속국가'].corr(network_log['정상/비정상'])
print('접속국가와 정상/비정상 열의 상관관계 :', corr)
corr = scaled_data['접속자'].corr(network_log['정상/비정상'])
print('접속자와 정상/비정상 열의 상관관계 :', corr)
corr = scaled_data['접속빈도'].corr(network_log['정상/비정상'])
print('접속빈도와 정상/비정상 열의 상관관계 :', corr)
```

```python
corr_matrix = pd.concat(objs=[scaled_data, network_log['정상/비정상']], axis=1).corr()
sns.heatmap(corr_matrix, annot=True)
```

접속자에 따라 정상/비정상 데이터가 강한 상관관계를 보이고, 접속국가, 아이피주소에 대해서는 어느정도의 상관관계를, 접속빈도에 대해서는 음의 상관관계를 갖네요!

```python
# 모델에 학습시키기 이전에 훈련세트와 테스트 세트에 대해서 나누어봅시다.
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data,
                                                                      target,
                                                                      random_state=42,
                                                                      test_size=0.2,
                                                                      stratify=target)
```

```python
# 이제 가장 먼저 로지스틱회귀로 해당 데이터를 훈련시켜보겠습니다.
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
lr = LogisticRegression(C=0.01, solver='liblinear')
pipeline = make_pipeline(trans,
                         lr)
pipeline.fit(train_input, train_target)
```

```python
pipeline.score(train_input, train_target)
```

```python
pipeline.score(test_input, test_target)
```
